# Pythonの基本的な数値計算
Python の基本的な数値計算をまとめました。足し算、引き算、かけ算、割り算の四則演算に加え、
剰余やべき乗の計算もデフォルトで用意されています。高度な計算はパッケージを使います。

## 足し算

```python
a = 1 + 2
b = 5 + (-7)
c = 1 + 3 + 5 + 7
```

```python
print(a)
print(b)
print(c)
```

## 引き算

```python
a = 7 - 3
b = -2 - (-6)
c = 1.5 - 0.3
```

```python
print(a)
print(b)
print(c)
```

## かけ算

```python
a = 7 * 3
b = (-4) * 8
c = 2j * 5j
```

```python
print(a)
print(b)
print(c)
```

jは複素数の表記。iではない。

## わり算

```python
a = 15 / 3
b = (-1) / 2
c = 6 / 1.5
```

```python
print(a)
print(b)
print(c)
```

```python
print(type(a))
```

型が浮動小数点数。普通のわり算は割られた値が整数であっても小数で表記される。

#### 切り捨てるとき

```python
a = 7 // 3
b = 15 // 5
c = -3 // 2
```

```python
print(a)
print(b)
print(c)
```
